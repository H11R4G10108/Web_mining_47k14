```python
from textblob import TextBlob
import re
import emot
from underthesea import sent_tokenize
from underthesea import word_tokenize
import os
from bs4 import BeautifulSoup
from emot.emo_unicode import UNICODE_EMOJI 
from emot.emo_unicode import EMOTICONS_EMO
directory = 'D:\Data1'
txt_files = [f for f in os.listdir(directory) if f.endswith('.txt')]
def remove_tags(html):
    soup = BeautifulSoup(html, "html.parser")
    for data in soup(['style', 'script']):
        data.decompose()
    return ' '.join(soup.stripped_strings)
def converting_emojis(text):
    for x in EMOTICONS_EMO:
        text = text.replace(x, "_".join(EMOTICONS_EMO[x].replace(",","").replace(":","").split()))
    for x in UNICODE_EMOJI:
        text = text.replace(x, "_".join(UNICODE_EMOJI[x].replace(",","").replace(":","").split()))
    return text
filename=os.path.join('D:', 'vietnamese-stopwords.txt')    
with open(filename, 'w+',encoding='utf-8') as p:
            List_StopWords=p.read().split("\n") 
for file in txt_files:
    filename=os.path.join('D:\Data1', file)    
    with open(filename, 'r',encoding='utf-8') as f:         
            line=f.read()
    line=line.replace("\n","")  # Remove the newline command
    line=line.replace("_"," ")
    line=line.lower() # Convert text to lowercase
    line=re.sub(r'[^\w\s]','',line) # Remove punctuation
    line = re.sub("\d+", " ", line) # Remove number
    line = re.sub(r"[!_@#$[]()]'", "", line) # Remove character: !@#$[]()
    line=remove_tags(line)
    line=converting_emojis(line)
    line= " ".join(line for line in line.split() if line not in List_StopWords)
    line = sent_tokenize(line)
    for x in line:
        x=word_tokenize(x)
        str=''
        for x in x:
            str+=''+x+','
        print(str)
        filename=os.path.join('D:\Data_Preprocessing', file)    
        with open(filename, 'w+',encoding='utf-8') as f:
             f.write(str)
```
