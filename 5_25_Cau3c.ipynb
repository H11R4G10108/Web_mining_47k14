```python
import re
import emot
from underthesea import sent_tokenize
import os
from emot.emo_unicode import UNICODE_EMOJI 
from emot.emo_unicode import EMOTICONS_EMO
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
def converting_emojis(text):
        for x in EMOTICONS_EMO:
            text = text.replace(x, "_".join(EMOTICONS_EMO[x].replace(",","").replace(":","").split()))
        for x in UNICODE_EMOJI:
            text = text.replace(x, "_".join(UNICODE_EMOJI[x].replace(",","").replace(":","").split()))
        return text
filename=os.path.join('D:', 'vietnamese-stopwords.txt')    
with open(filename, 'r',encoding='utf-8') as p:
                List_StopWords=p.read().split("\n") 
directory = 'D:\Data1'
txt_files = [f for f in os.listdir(directory) if f.endswith('.txt')]
Data=''
for file in txt_files:
        filename=os.path.join('D:\Data1', file)    
        with open(filename, 'r',encoding='utf-8') as f:
            text=f.read()
        Data=Data+','+text
Data=Data.replace("\n","")  # Remove the newData command
Data=Data.replace("_"," ")
Data=Data.lower() # Convert text to lowercase
Data=re.sub(r'[^\w\s]','',Data) # Remove punctuation
Data = re.sub("\d+", " ", Data) # Remove number
Data = re.sub(r"[!_@#$[]()]'", "", Data) # Remove character: !@#$[]()
Data=converting_emojis(Data)
Data= " ".join(Data for Data in Data.split() if Data not in List_StopWords)
Data = sent_tokenize(Data)
for i in range (0, len(Data)):
    Data[i]=re.sub(r'[^\w\s]','',Data[i])
tr_idf_model  = TfidfVectorizer()
tf_idf_vector = tr_idf_model.fit_transform(Data)
tf_idf_array = tf_idf_vector.toarray()
# print(tf_idf_array)
W = tr_idf_model.get_feature_names_out()# print(W)
df_tf_idf = pd.DataFrame(tf_idf_array, columns = W)
pd.set_option('display.max_column', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_seq_items', None)
pd.set_option('display.max_colwidth', 500)
pd.set_option('expand_frame_repr', True)
print(df_tf_idf)
```
